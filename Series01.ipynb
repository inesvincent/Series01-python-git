## Minimal script to see the ECG signal

```python
# load the libraries
import numpy as np
import matplotlib.pyplot as plt

#load the data
ecg_raw = np.loadtxt('data/ECGu.txt')
print(ecg_raw.shape)
print(ecg_raw)
```

```python
# plot the data
plt.plot(ecg_raw)
plt.title('ECG signal')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

```python
# create t the time vector in seconds
sampling_frequency = 1000  # in Hz
T = 1 / sampling_frequency  # Sampling period in seconds
N = len(ecg_raw)  # Number of samples (nb mesure prise)
start = 0  # start time in seconds
stop = (N - 1) * T  # end time in seconds

# time vector
t = np.linspace(start, stop, N)
print(t.shape)
print(t)

#plot the data with time vector
plt.plot(t, ecg_raw)
plt.title('ECG signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

```python
# Rescale the ecg signal to mV knowing that 1 unit = 1024 microV
ecg = ecg_raw * 1024 * 1e-6 # in mV A/D converter gain : 1024 V per unit  

#plot the ecg signal in mV
plt.plot(t, ecg)
plt.title('ECG signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (mV)')
plt.grid()
plt.show()
```
